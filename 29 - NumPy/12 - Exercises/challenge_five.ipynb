```python
import csv
import numpy as np
from dateutil import parser
import pandas as pd

# Clues
# Sweatshirt from Noahs Market (find produc/product code/orders/customers)
# Lives in Staten Island

# Customers from Staten Island

# Staten Island, NY

with open("noahs-customers.csv") as custs:
    reader = csv.reader(custs)
    headers = next(reader)
    data = list(reader)
    
cust_arr = np.array(data)
cities = cust_arr[:, 3]

staten_chars = [city[0:17] for city in cities]
staten_char_arr = np.array(staten_chars)
only_statens_mask = np.equal(staten_char_arr, "Staten Island, NY")

only_staten_custs = cust_arr[only_statens_mask]

cust_series = pd.Series(only_staten_custs[:, 5], index = only_staten_custs[:, 0])
    
# Sweatshirt Orders
# Noah's Jersey

# All Jersey skus
    
with open("noahs-products.csv") as prods:
    reader = csv.reader(prods)
    headers = next(reader)
    data = list(reader)
    
prods_arr = np.array(data)
descs = prods_arr[:, 1]
    
parsed_descs = [desc[0:13] for desc in descs]
parsed_descs_arr = np.array(parsed_descs)

only_jersey_mask = np.equal(parsed_descs_arr, "Noah's Jersey")

only_jersey_prods = prods_arr[only_jersey_mask]
only_jersey_skus = only_jersey_prods[:, 0]
# print(only_jersey_skus)

# Order w jerseys

with open("noahs-orders_items.csv") as ord_items:
    reader = csv.reader(ord_items)
    headers = next(reader)
    data = list(reader)
    
ord_item_arr = np.array(data)
order_skus = ord_item_arr[:, 1]
sku_letters = [ item[0:3] for item in order_skus]
mask = np.equal(np.array(sku_letters), "COL")
only_cols = ord_item_arr[mask]
ord_ids = only_cols[:, 0]

# Create keys for orders

with open("noahs-orders.csv") as ords:
    reader = csv.reader(ords)
    headers = next(reader)
    data = list(reader)
    
ords_arr = np.array(data)

ords_series = pd.Series(ords_arr[:, 1], index = ords_arr[:, 0])

cust_list = []

for order in ord_ids:
        cust_list.append(ords_series[order])
        
poss_cust_nums = []

for cust in cust_list:
    try:
        poss_cust_nums.append(cust_series[cust])
    except:
        pass
    
print(set(poss_cust_nums))
        

```
