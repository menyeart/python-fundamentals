```python
import csv
import numpy as np
from dateutil import parser
import pandas as pd

# Customers from Staten Island

with open("noahs-customers.csv") as custs:
    reader = csv.reader(custs)
    headers = next(reader)
    data = list(reader)
    
cust_arr = np.array(data)
# cust_data_frame = pd.DataFrame(cust_arr, index = cust_arr[:, 3], columns = headers )
# staten_custs = cust_data_frame.loc['Staten Island, NY 10304']
# print(staten_custs.loc[:, ['name', 'phone']])


cities = cust_arr[:, 3]

staten_chars = [city[0:17] for city in cities]
staten_char_arr = np.array(staten_chars)
only_statens_mask = np.equal(staten_char_arr, "Staten Island, NY")

only_staten_custs = cust_arr[only_statens_mask]

cust_series = pd.Series(only_staten_custs[:, 5], index = only_staten_custs[:, 0])
    
# Cat Orders

# # All cat skus
    
with open("noahs-products.csv") as prods:
    reader = csv.reader(prods)
    headers = next(reader)
    data = list(reader)
    
prods_arr = np.array(data)

skus = prods_arr[:, 0]
parsed_skus = [desc[0:3] for desc in skus]
parsed_skus_arr = np.array(parsed_skus)

only_pet_mask = np.equal(parsed_skus_arr, "PET")
only_pet_prods = prods_arr[only_pet_mask]

# print(only_pet_prods)

pet_prod_descs = only_pet_prods[0:, 1]

senior_cat_foods = [ 
    True if "Senior" in desc and "Cat" in desc else False for desc in pet_prod_descs
        ]
old_cat_foods = only_pet_prods[senior_cat_foods]
cat_food_skus =old_cat_foods[:, 0]



with open("noahs-orders_items.csv") as ord_items:
    reader = csv.reader(ord_items)
    headers = next(reader)
    data = list(reader)
    
ord_items_arr = np.array(data)

# ord_items_frame = pd.DataFrame(ord_items_arr, index = ord_items_arr[:,1], columns = headers)
ord_items_series = pd.Series(ord_items_arr[:, 0], index = ord_items_arr[:, 1])

cat_items = [ ord_items_series.loc[sku].values for sku in cat_food_skus ] 

with open("noahs-orders.csv") as ords:
    reader = csv.reader(ords)
    headers = next(reader)
    data = list(reader)
    
ord_arr = np.array(data)
ord_series = pd.Series(ord_arr[:, 1], index = ord_arr[:, 0])

# print(ord_series)

cust_ids_cats = [ list(ord_series[item]) for item in cat_items ]

ids = [cat for cat in cust_ids_cats]

print(ids)
# Create a dataframe

# cust_arr = np.array(data)

# cust_frame = pd.DataFrame(cust_arr, index = cust_arr[:, 0], columns = headers)
# # print(cust_frame)
# print(cust_frame.loc['1001','name'])
```
